## **Function to derive day of DOB**

This function deals with taking the **date of birth** as input and deriving the day of DOB from it.



```python
# Function to find the day of week, given the date

import calendar
import datetime

def dayDate(date):
  dob = datetime.datetime.strptime(date, '%d %m %Y').weekday()
  l = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

  return (l.index(calendar.day_name[dob]) + 1)
```

## **Function to generate a Uniform PDF**

```python
from scipy.stats import uniform
import numpy as np

def uniformDistribution(day, month):

  a = min(day, month)
  b = max(day, month)

  x1 = np.linspace(a, b, 500)
  x2 = np.arange(0, a, 0.01)
  x3 = np.arange(b, 30, 0.01)

  y1 = np.random.uniform(a, b, size = (500,))
  y2 = np.array([0] * (a * 100))
  y3 = np.array([0] * ((30 - b) * 100))

  x = np.concatenate([x2, x1, x3])
  y = np.concatenate([y2, y1, y3])

  print("500 random numbers in a uniform distribution:\n", y1)

  return x, y, y1
```

# **Function to generate Gaussian PDF**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def GaussianDistribution(dayNo, month):
  y = np.random.normal(loc = dayNo, scale = (month / 4), size = 500)
  print("500 random numbers in a normal distribution:\n", y)

  return y
```
