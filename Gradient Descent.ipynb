### **Question 1**

Minimize the given function f(x) using Gradient Descent with different learning rates.

```python
import numpy as np
import matplotlib.pyplot as plt

# Compute the value of f(x)
def f1(x):
  return ((x ** 4) - (8 * (x ** 2))) 

# Compute the value of f'(x)
def dfx1(x):
  return (((4 * (x ** 3)) - (16 * x)))

# Create arrays for x and corresponding y values to plot given f(x)
x2 = np.arange(-2.5, 2.5, 0.001)
y2 = np.array([f1(i) for i in x2])

# Question 1(a): Plot the given function
plt.figure(figsize = [12, 6])
plt.grid()
plt.xlabel("x")
plt.ylabel("y = (x ** 4) - (8 * (x ** 2))")
plt.title("Function for varying learning rates")

# Question 1(b): Minimize f(x) with varying learning rates
learning_rates = [(0.002 * i) for i in range(1, 5)]

for lr in learning_rates:
  x = 0.5
  f_init = [1000]
  x_init = [0]

  fn = f1(x)
  i = 0
  print("===================================================================\n")

  # While the function is decreasing, update the value of x
  while (f_init[-1] > fn):
    i += 1
    xold = x
    diffx = dfx1(x)

    update = lr * diffx

    # Check for vanishing gradient
    if (update == 0):
      print("Gradient vanished [f'(x) = 0].\n")
      f_init.append(fn)
      fn = f1(x)
      x_init.append(x)
    else:
      x -= update
      x_init.append(x)
      f_init.append(fn)
      fn = f1(x)
    
    print("Learning Rate: ", lr)
    print("\nIteration ", i, ":\n")
    print("x", (i + 1), " = ", "x", i, " - (learning rate) * df/dx \n")
    print("x", (i + 1), " = ", xold, " - ", lr, "*", diffx, "\n")
    print("x", (i + 1), " = ", x, "\n\n")
    print("-------------------------------------------------------------------")

  f_real = [0] + f_init[1:]
  plt.plot(x2, y2)
  str1 = "LR: " + str(lr)
  plt.plot(x_init, f_real, label = str1)
```

### **Question 2**

Find x and y which minimize the given function f(x, y) using Gradient Descent and learning rates between 0 and 1.

```python
import matplotlib.pyplot as plt
import numpy as np

# Compute the value of f(x, y)
def f(x, y):
  return (((x ** 2) + y - 11) ** 2) + ((x + (y ** 2) - 7) ** 2) 

# Compute the partial derivative of f(x, y) w.r.t. x
def dfx(x, y):
  return (4 * x * ((x ** 2) + y - 11)) + (2 * (x + (y ** 2) - 7))

# Compute the partial derivative of f(x, y) w.r.t. y
def dfy(x, y):
  return 2 * ((x ** 2) + y - 11) + (4 * y * (x + (y ** 2) - 7))

# Base coordinates
[x, y] = [0.07, 0.11]

f_init = [1000]
lr = 0.05

fn = f(x, y)
i = 0
while (f_init[-1] >= fn):
  i += 1
  f_init.append(fn)
  xold = x
  yold = y
  diffx = dfx(x, y)
  diffy = dfy(x, y)

  x -= lr * diffx
  y -= lr * diffy

  fn = f(x, y)
  print("\nIteration ", i, ":\n")
  print("x", (i + 1), " = ", "x", i, " - (learning rate) * df/dx \n")
  print("x", (i + 1), " = ", xold, " - ", lr, "*", diffx, "\n")
  print("x", (i + 1), " = ", x, "\n\n")

  
  print("y", (i + 1), " = ", "y", i, " - (learning rate) * df/dy \n")
  print("y", (i + 1), " = ", yold, " - ", lr, "*", diffy, "\n")
  print("y", (i + 1), " = ", y, "\n\n")
  print("---------------------------------------------------------------------")

print("The optimal values of x and y are:\n")
print("x: ", x)
print("y: ", y)
```
